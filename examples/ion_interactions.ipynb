```python
import pandas as pd
pd.set_option("display.max_columns",999) # display all columns

from pymatgen.analysis.local_env import BrunnerNN_reciprocal, BrunnerNN_relative, BrunnerNN_real, EconNN, JMolNN, \
                                        MinimumDistanceNN, MinimumOKeeffeNN, MinimumVIRENN, \
                                        VoronoiNN, VoronoiNN_modified, CrystalNN
from materialscoord.core import Benchmark, HumanInterpreter

methods = [BrunnerNN_reciprocal(), BrunnerNN_relative(), BrunnerNN_real(), EconNN(), JMolNN(),
           MinimumDistanceNN(), MinimumOKeeffeNN(), MinimumVIRENN(), VoronoiNN(), VoronoiNN_modified(),
           HumanInterpreter(), CrystalNN()]

structure_groups = ["common_binaries", "zeolites", "mixed", "elemental"] 

unique_sites = 13

algo = ["BrunnerNN_reciprocal", "BrunnerNN_relative", "BrunnerNN_real", "CrystalNN", "EconNN", "JMolNN",
        "MinimumDistanceNN", "MinimumOKeeffeNN", "MinimumVIRENN", "VoronoiNN", "VoronoiNN_modified"]
```

```python
cat_df = Benchmark(methods=methods, structure_groups=structure_groups, unique_sites=True,
                  use_weights=False, cations=True)
cat_df.benchmark()

cats = cat_df.report(totals=False, separate_columns=True, max_sites=unique_sites)
cats
```

```python
from materialscoord.nb_funcs import nb_funcs # used to calculate algo accuracy scores
nbf_cats = nb_funcs(cats, algo, unique_sites=unique_sites, cations=True)
```

```python
final_cats = nbf_cats.final()
final_cats.loc['total'] = final_cats.sum(axis=0)
final_cats
```

```python
methods = [BrunnerNN_reciprocal(), BrunnerNN_relative(), BrunnerNN_real(), EconNN(), JMolNN(),
           MinimumDistanceNN(), MinimumOKeeffeNN(), MinimumVIRENN(), VoronoiNN(), VoronoiNN_modified(),
           HumanInterpreter(), CrystalNN()]

structure_groups = ["common_binaries", "zeolites", "mixed"] 

algo = ["BrunnerNN_reciprocal", "BrunnerNN_relative", "BrunnerNN_real", "CrystalNN", "EconNN", "JMolNN",
        "MinimumDistanceNN", "MinimumOKeeffeNN", "MinimumVIRENN", "VoronoiNN", "VoronoiNN_modified"]

an_df = Benchmark(methods=methods, structure_groups=structure_groups, unique_sites=True,
                  use_weights=False, anions=True)
an_df.benchmark()

ans = an_df.report(totals=False, separate_columns=True, max_sites=unique_sites)
ans
```

```python
nbf_ans = nb_funcs(ans, algo, unique_sites=unique_sites, cations=False, anions=True)
```

```python
final_ans = nbf_ans.final()
final_ans.loc['total'] = final_ans.sum(axis=0)
final_ans 
```

```python
final = final_cats.add(final_ans, fill_value=0)
final
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20, 10))

sns.set(font='Times New Roman')
sns.set(font_scale=1)

hm = sns.heatmap(final, annot=True, cmap="BuPu", vmax=10)

ax.set_xticklabels(final.columns.tolist(), rotation=60)
ax.set_yticklabels(final.index.tolist())

plt.show()
```
