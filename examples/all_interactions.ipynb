```python
import pandas as pd
pd.set_option("display.max_columns",999) # display all columns

from pymatgen.analysis.local_env import BrunnerNN_reciprocal, BrunnerNN_relative, BrunnerNN_real, EconNN, JMolNN, \
                                        MinimumDistanceNN, MinimumOKeeffeNN, MinimumVIRENN, \
                                        VoronoiNN, VoronoiNN_modified, CrystalNN
from materialscoord.core import Benchmark, HumanInterpreter, NbFuncs

methods = [BrunnerNN_reciprocal(), EconNN(), JMolNN(), MinimumDistanceNN(), MinimumOKeeffeNN(), MinimumVIRENN(),
           VoronoiNN(tol=0.5), CrystalNN(), HumanInterpreter()]
#structure_groups = ["A2BX4", "ABX3", "ABX4", "common_binaries", "elemental"]
structure_groups = ["A2BX4", "ABX3", "ABX4"]

cat_bm = Benchmark(methods=methods, structure_groups=structure_groups, unique_sites=True,
               use_weights=False, cation_anion=True, anion_cation=False)
cat_bm.benchmark()

report = cat_bm.report()
report

cat_nbf = NbFuncs(cat_bm)
cat = cat_nbf.final()
cat
```

```python
an_bm = Benchmark(methods=methods, structure_groups=structure_groups, unique_sites=True,
               use_weights=False, cation_anion=False, anion_cation=True)
an_bm.benchmark()

report = an_bm.report()
report

an_nbf = NbFuncs(an_bm)
an = an_nbf.final()
an = an.fillna(0)
an
```

```python
df = cat.add(an)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20, 15))

sns.set(font='Times New Roman')
sns.set(font_scale=1)

hm = sns.heatmap(df, annot=True, cmap="BuPu", vmax=10)

ax.set_xticklabels(df.columns.tolist(), rotation=60)
ax.set_yticklabels(df.index.tolist())

plt.savefig('all_interactions.png')
plt.show()
```
