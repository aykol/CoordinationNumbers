# MaterialsCoord benchmarking – tuning the `EconNN` tolerance parameter

The performance of EconNN can be tuned through the `tol` parameter that determines which neighbors are considered bonded. By default, in MaterialsCoord we have used a `tol` value of 0.2, however, this might not be the optimum. In those notebook, we trial a number of tolerance parameters and compare the results.

*Written using:*
- MaterialsCoord==0.2.0

*Authors: Alex Ganose (03/30/20)*

---

First, lets initialize a series of `EconNN` classes with difference tolerance parameters.

```python
import numpy as np

from pymatgen.analysis.local_env import EconNN

tolerances = np.linspace(0.05, 1, 21)
nn_methods = [EconNN(tol=t) for t in tolerances]
```

Next, import the benchmark and choose which structures we are interested in.

```python
from materialscoord.core import Benchmark

structure_groups = ["common_binaries", "elemental", "A2BX4", "ABX3", "ABX4"]

bm = Benchmark.from_structure_group(structure_groups)
```

Calculate the scores.

```python
scores = bm.score(nn_methods)
```

Plot the tolerance value against benchmark score.

```python
import matplotlib.pyplot as plt
from matplotlib import ticker
from pathlib import Path
import seaborn as sns

totals = scores.loc["Total"]

# define matplotlib style settings
style = {
    "font.sans-serif": ["Helvetica", "Arial"], "axes.labelsize": 16,
    "xtick.labelsize": 16, "ytick.labelsize": 16, "xtick.direction": "in",
    "ytick.direction": "in", "xtick.major.size": 8, "xtick.minor.size": 4,
    "ytick.major.size": 8, "ytick.minor.size": 4, "lines.linewidth": 2.5,
    "lines.markersize": 10, "axes.linewidth": 1.2, "xtick.major.width": 1.2,
    "xtick.minor.width": 1.2, "ytick.major.width": 1.2, "ytick.minor.width": 1.2, 
    "pdf.fonttype": 42
}

plt.style.use(style)
fig, ax = plt.subplots(figsize=(6, 6))

ax.plot(tolerances, totals)
ax.set(ylabel="Benchmark score", xlabel="Tolerance", xlim=(0.05, 1))

plt.savefig(Path("plots", "EconNN-tolerance-tuning.pdf"), bbox_inches="tight")
plt.show()
```
