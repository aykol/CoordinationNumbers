```python
import pandas as pd
pd.set_option("display.max_columns",999) # display all columns
pd.set_option("display.max_rows", 999)

from pymatgen.analysis.local_env import BrunnerNN_reciprocal, EconNN, JmolNN, \
                                        MinimumDistanceNN, MinimumOKeeffeNN, MinimumVIRENN, \
                                        VoronoiNN, VoronoiNN_modified, CrystalNN
from materialscoord.core import Benchmark, HumanInterpreter

methods = [MinimumDistanceNN(),  MinimumOKeeffeNN(), MinimumVIRENN(), JmolNN(), EconNN(), 
           BrunnerNN_reciprocal(), VoronoiNN(tol=0.5), CrystalNN(), HumanInterpreter()]
structure_groups = ["A2BX4", "ABX3", "ABX4"]

preset = Benchmark.from_preset(preset_name=structure_groups)
preset.benchmark(methods=methods)
```

```python
df = preset.score(methods)
df
```

```python
print(df.index.tolist())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize=(17, 14))

sns.set(font='Times New Roman')
sns.set(font_scale=1.5)

hm = sns.heatmap(df, annot=True, cmap="Blues", vmax=10)

strucs = [r'AlAsO$_4$', r'BAsO$_4$', r'BaAl$_2$O$_4$', r'BaTiO$_3$', r'BaZnF$_4$', r'Be$_2$SiO$_4$', 
          r'CaB$_2$O$_4$', r'CaCO$_3$ (aragonite)', r'CaCO$_3$ (calcite)', r'CaFe$_2$O$_4$', r'CaWO$_4$',
          r'CePO$_4$', r'CrVO$_4$', r'Fe$_2$SiO$_4$', r'FeTiO$_3$', r'FeWO$_4$', r'GaPO$_4$', r'K$_2$NiF$_4$',
          r'K$_2$SO$_4$', r'MgAl$_2$O$_4$', r'MgUO$_4$', r'MnMoO$_4$', r'Na$_2$SO$_4$', r'SbNbO$_4$', 
          r'Sr$_2$PbO$_4$', r'SrTiO$_3$', r'TlAlF$_4$', r'ZnSO$_4$', r'ZrSiO$_4$', 'Total']
          
ax.set_yticklabels(strucs)

nn = ["MinimumDistanceNN", "MinimumOKeeffeNN", "MinimumVIRENN", "JmolNN", 
      "EconNN", "BrunnerNN", "VoronoiNN", "CrystalNN"]
ax.set_xticklabels(nn, rotation=60)

cbar = ax.collections[0].colorbar
cbar.set_label('Benchmark Score', labelpad=40, rotation=270)

ax.axhline(y=0, color='k',linewidth=1)
ax.axhline(y=len(strucs), color='k',linewidth=2)
ax.axhline(y=len(strucs)-1, color='k',linewidth=1)
ax.axvline(x=0, color='k',linewidth=1)
ax.axvline(x=len(nn), color='k',linewidth=2)

plt.savefig('ternary_cation.pdf', bbox_inches='tight')
plt.show()
```
